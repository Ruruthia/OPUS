```python
import pickle
import pandas as pd
from funk_svd import SVD
```

```python
df = pd.read_csv('/pio/scratch/1/recommender_systems/raw/Amazon/Clothing_Shoes_and_Jewelry.csv',
                 names=['asin', 'reviewerID', 'overall', 'unixReviewTime'])
```

```python
df = df.rename(columns={"reviewerID": "u_id", "asin": "i_id", "overall": "rating"})
```

```python
model = SVD(lr=0.001, reg=0.005, n_epochs=300, n_factors=100, min_rating=1, max_rating=5)
```

```python
model.fit(df)
```

```python
pickle.dump(model, open('/pio/scratch/1/i313924/SVD_data/model_300_epochs.pkl', 'wb'), protocol=4)
```
