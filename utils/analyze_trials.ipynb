```python
import pickle

import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "svg"

with open('/pio/scratch/1/i313924/data/lightfm_data/hyperopt_trials_adadelta.pickle', 'rb') as f_in:
    trials = pickle.load(f_in)

results_pdf = pd.DataFrame([
    {
        'hit_rate': -trial['result']['loss'],
        "no_components": trial['misc']['vals']["no_components"][0],
        "rho": trial['misc']['vals']["rho"][0],
        "epsilon": trial['misc']['vals']["epsilon"][0],
        "item_alpha": trial['misc']['vals']["item_alpha"][0],
        "user_alpha": trial['misc']['vals']["user_alpha"][0],
        "max_sampled": trial['misc']['vals']["max_samples"][0],
    }
    for trial in trials.trials
])
results_pdf.sort_values('hit_rate', ascending=False)
```

```python
px.parallel_coordinates(results_pdf, color='hit_rate')
```
